#### Make File Links

```python
base_url = 'https://github.com/gumdropsteve/intro_to_python/raw/main/day_10/data/'

locations = ['las_vegas']

filters = ['', 'entire_homes', 'private_rooms', 'hotel_rooms', 
           'super_hosts', 'entire_home_super_hosts', 'private_room_super_hosts', 'hotel_room_super_hosts']

dates = ['2021-01-09', '2021-01-10', '2021-01-11', '2021-01-12', '2021-01-13']

urls = []
for l in locations:
    for f in filters:
        for d in dates:
            if f != '':
                urls.append(f'{base_url}{l}_{f}_{d}.xls')
            else:
                urls.append(f'{base_url}{l}{f}_{d}.xls')
```

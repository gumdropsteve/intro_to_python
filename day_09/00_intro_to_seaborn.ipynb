#### pairwise data relationships
https://seaborn.pydata.org/tutorial/axis_grids.html

```python
import seaborn as sns

# load in iris datset from seaborn
iris = sns.load_dataset("iris")

# make a parigrid
g = sns.PairGrid(iris)

# plot scatter plots
g.map(sns.scatterplot)
```

```python
# load in iris datset from seaborn
iris = sns.load_dataset("iris")

# make a parigrid
g = sns.PairGrid(iris, hue='species')

# plot scatter plots
g.map(sns.scatterplot)
```

## Using Random Data (blobs.csv)

```python
import pandas as pd

blobs = pd.read_csv('https://github.com/gumdropsteve/datasets/raw/master/blobs.csv')

blobs
```

```python
# above takes too long to run... too much data for my system with seaborn
blobs_small = blobs.sample(1000)

del blobs
```

#### Is there a time difference between adding a hue and not adding a hue

```python
%%time
# pick the columns I want to focus on
blobs_small_2 = blobs_small[['a', 'b', 'c', 'd']]

# make the parigred
g = sns.PairGrid(blobs_small_2)

# apply the scatter plots
g.map(sns.scatterplot)
```

```python
blobs_adjusted = blobs_small[['a', 'b', 'c', 'd', 'Y']]

blobs_adjusted['Y'] = blobs_adjusted['Y'].astype('str')
```

```python
%%time
# pick the columns I want to focus on
blobs_small_2 = blobs_adjusted[['a', 'b', 'c', 'd', 'Y']]

# make the parigred
g = sns.PairGrid(blobs_small_2, hue='Y')

# apply the scatter plots
g.map(sns.scatterplot)
```

## You can do more with PairGrid

```python
# with iris
g = sns.PairGrid(iris)

g.map_diag(sns.histplot)

g.map_offdiag(sns.scatterplot)
```

```python
%%time
# with iris
g = sns.PairGrid(iris, hue='species')

g.map_diag(sns.histplot)

g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(blobs_adjusted, hue='Y')

g.map_diag(sns.histplot)

g.map_offdiag(sns.scatterplot)
```
